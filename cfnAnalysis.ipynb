# CFN Street Fighter 6 Analysis 

```python
import numpy
import matplotlib as plt
import pandas as pd
```

After importing all that we need, let's read the csv tables that we have received previously.

```python
players = pd.read_csv("master.csv", names = ["Player", "Character", "Profile", "Country", "LP"])
players
```

We have now placed our csv in a pandas dataframe. This current data only has about 2000 entries, but we can get more from our CFN list. For now, let's analyze this data. Let's see what country is featured the most in the top 2000.

```python
country = players["Country"].value_counts().plot(kind="bar", title ="Most Popular Country", figsize=(15, 10), legend=True, fontsize=12)
country.set_ylabel("Frequency")
country
```

It seems that Japan takes the top spot in the top 2000 players, as expected. Now, let's take a look at the characters most prominently featured in the top 2000. 

```python
character = players["Character"].value_counts().plot(kind="bar")
character
```

Ken is still reigned supreme even among the highest players. Now, let's take a look at the description of LP statistics

```python
description = players["LP"].describe()
description
```
